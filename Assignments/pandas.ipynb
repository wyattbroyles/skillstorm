```python
#Using the imdb top 1000 movies dataset find the following
import pandas as pd

df1 = pd.read_csv("../Etc/imdb_top_1000.csv")
df1.head(20)
```

```python
#Get rid of useless columns
df1["Released_Year"].value_counts()
df1.loc[df1["Released_Year"] == "PG"]
df1.loc[df1["Series_Title"] == "Apollo 13"]
df1.at[966, "Released_Year"] = 1995
df1.at[966, "Released_Year"]
```

```python
# oldest movie
oldest_year = df1['Released_Year'].min()
print(oldest_year)
```

```python
#newest movie
newest_year = df1['Released_Year'].max()
print(newest_year)
```

```python
# top 10 movies based on rating
df1_sorted = df1.sort_values(by= 'IMDB_Rating', ascending=False )

newest_movies = df1_sorted.head(10)
print(newest_movies["IMDB_Rating"])
```

```python
#Top movie for each genre

unique_genres = df1["Genre"].unique()

for genre in unique_genres:
    print("GENRE: ",genre)
    movies_in_genre = df1[df1["Genre"] == genre]
    df1_sorted = movies_in_genre.sort_values(by= 'IMDB_Rating', ascending=False)
    top_movie = df1_sorted.head(1)
    print(top_movie)
```

```python
#Director with the most movies

directors = df1["Director"].value_counts()
max_count = directors.max()
most_frequent_director = directors.idxmax()
print(most_frequent_director,max_count)
```

```python
#Star with the most movies

stars1 = df1["Star1"].value_counts()
most_frequent_stars1 = stars1.idxmax()
print(most_frequent_stars1,stars1.max())
```

```python
# Highest grossing movie for each genre

unique_genres = df1["Genre"].unique()

for genre in unique_genres:
    movies_in_genre = df1[df1["Genre"] == genre]
    df1_sorted = movies_in_genre.sort_values(by= 'Gross', ascending=False )
    top_gross = df1_sorted["Gross"].head(1)
    genre_movie = df1_sorted["Genre"].head(1)
    print(genre_movie, top_gross)
```

```python
#Lowest grossing movie for each director

unique_directors = df1["Director"].unique()

for director in unique_directors:
    movies_by_director = df1[df1["Director"] == director]
    df1_sorted = movies_by_director.sort_values(by= 'Gross')
    low_gross = df1_sorted["Gross"].head(1)
    print(director, low_gross)
```

```python
#Aggregate the gross revenue for all 1000 movies.

df2 = pd.read_csv("../Etc/imdb_top_1000.csv")
df2['Gross'] = df2['Gross'].str.replace(',', '')
df2['Gross'] = pd.to_numeric(df2['Gross'])
df2['Gross'] = df2['Gross'].fillna(0)
total_gross = df2["Gross"].sum()
print(f"${total_gross:,.0f}")
```
